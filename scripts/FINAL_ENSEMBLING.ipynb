```python
from collections import Counter

# implementing multimode
def multimode(lst):
    res = []
    test_list1 = Counter(lst) 
    temp = test_list1.most_common(1)[0][1] 
    for ele in lst:
        if lst.count(ele) == temp:
            res.append(ele)
            
    res = list(set(res))
    
    return res
```
